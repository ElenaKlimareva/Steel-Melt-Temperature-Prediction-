Steel Melt Temperature Prediction (Industrial Optimization)
</h1>

**Project Description:**

**Client:** a steelworks.

To optimise production costs, the client aims to reduce electricity consumption at the steel processing stage. To achieve this, the temperature of the alloy must be controlled.


**Project Objective:** to build a model that predicts the temperature of the alloy.

The client plans to use the developed model to simulate the technological process.


**Description of the Technological Process:**

Steel is processed in a metal ladle with a capacity of around 100 tonnes. To allow the ladle to withstand high temperatures, its inner surface is lined with refractory brick. Molten steel is poured into the ladle and heated to the required temperature using graphite electrodes installed in the ladle lid.

Sulphur is removed from the alloy (this process is desulphurisation), impurities are added to adjust the chemical composition, and samples are taken. The steel is alloyed — its composition is modified — by feeding alloy pieces from a bulk materials hopper or wire through a special trib-apparatus (*tribe* — “mass”).

Before alloying additives are introduced for the first time, the steel temperature is measured and its chemical analysis is performed. Then the temperature is increased for several minutes, alloying materials are added, and the alloy is purged with an inert gas. After that, it is stirred and measurements are taken again. This cycle is repeated until the target chemical composition and optimal melting temperature are achieved.

The molten steel is then sent for secondary metallurgy treatment or fed into a continuous casting machine. From there, the finished product emerges in the form of slab billets (*slab* — “plate”).


**Data Description**

The data consist of several files obtained from different sources:

- `data_arc_new.csv` — electrode data;
- `data_bulk_new.csv` — bulk material feed data (volume);
- `data_bulk_time_new.csv` — bulk material feed data (time);
- `data_gas_new.csv` — gas purging data;
- `data_temp_new.csv` — temperature measurement results;
- `data_wire_new.csv` — wire material data (volume);
- `data_wire_time_new.csv` — wire material data (time).

In all files, the `key` column contains the batch number. There may be multiple rows with the same `key` value in the files; these correspond to different processing iterations.

The time features `arc_heating_start` and `arc_heating_end` do not themselves carry useful information for predicting the target feature `temperature`, but they can be used to derive a potentially useful feature - `heating_duration`:

```python
data_arc['heating_duration_min'] = \
                    (data_arc['arc_heating_end'] - data_arc['arc_heating_start']).dt.total_seconds() / 60
```

From `active` and `reactive_power`, we will generate a potentially useful feature - `apparent_power`:

```python
data_arc['apparent_power'] = \
                    (data_arc['active_power']**2 + data_arc['reactive_power']**2)**0.5
```

```python
data_arc.head()
```

Let's aggregate the dataset by `key`, as `key` is the object of modelling:

```python
data_arc_final = data_arc.groupby("key").agg({
    "active_power": "sum",
    "reactive_power": "sum",
    "apparent_power": "sum",
    "heating_duration_min": "sum"
}).reset_index()
```

**Let's plot histograms and boxplots for the numerical features in our dataset:**

```python
def plot_hist(data, parameter, title):
    plt.figure(figsize=(5.5, 3.8))

    ax = sns.histplot(
        data=data,
        x=parameter,
        stat='probability',
        common_norm=False,
        bins='auto',
        kde=False,
        alpha=0.7
    )

    plt.axvline(data[parameter].mean(), color='black', linestyle='-', label='Mean')
    plt.axvline(data[parameter].median(), color='black', linestyle='--', label='Median')

    
    legend_patches = [
        mlines.Line2D([], [], color='black', linestyle='-', label='Mean'),
        mlines.Line2D([], [], color='black', linestyle='--', label='Median')
    ]

    plt.legend(handles=legend_patches)
    
    
    plt.title(title, fontsize=10)
    plt.xlabel(parameter)
    plt.ylabel('Share')
    plt.tight_layout()
    plt.show()
```

```python
def boxplot(data, parameter, title):
    plt.figure(figsize=(5.5, 3.8))
    sns.boxplot(data=data[parameter])
    plt.title(title, fontsize=10)
    plt.ylabel(parameter)
    plt.tight_layout()
    plt.show()
```

```python
def plot_hist(data, column, title, ax):
    ax.hist(data[column], bins=50, edgecolor='black', alpha=0.7)
    ax.set_title(title)
    ax.set_xlabel(column)
    ax.set_ylabel('Count')


def boxplot(data, column, title, ax):
    ax.boxplot(data[column], vert=False)
    ax.set_title(title)
    ax.set_xlabel(column)
```

## Data preparation

```python
RANDOM_STATE = 21225
TEST_SIZE = 0.25
TARGET = 'final_temperature'
```

```python
X = data_final.drop(columns=[TARGET, 'key'])
y = data_final[TARGET]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE
)
```

```python
num_cols = X.select_dtypes(include=['number']).columns.tolist()
```

```python
num_pipe = Pipeline([
    ('scaler', MinMaxScaler())
])


preprocessor = ColumnTransformer([
    ('num', num_pipe, num_cols)
])


pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('feature_selection', SelectKBest(f_regression, k='all')),
    ('models', LinearRegression())
])
```

## Machine learning model training

```python
param_distributions_rf = [
    

    {
        'models': [RandomForestRegressor(random_state=RANDOM_STATE)],
        'models__n_estimators': [200, 400, 600, 800],
        'models__max_depth': [None, 10, 20, 30, 50],
        'models__min_samples_split': [2, 5, 10],
        'models__min_samples_leaf': [1, 2, 4],
        'models__max_features': ['sqrt', 'log2', 0.8]
    }]
```

```python
search_rf = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=param_distributions_rf,
    n_iter=20,
    scoring='neg_mean_absolute_error',
    cv=5,
    n_jobs=-1,
    random_state=RANDOM_STATE,
    refit=True,
    verbose=1
)
```

```python
search_rf.fit(X_train, y_train)

print('Random Forest model and its parameters:\n', search_rf.best_estimator_)
print('CV MAE (RandomForest):', round(-search_rf.best_score_, 2))
```

```python
param_distributions_cat_boost = [
    
    {
        'models': [
            CatBoostRegressor(
                loss_function='MAE',
                verbose=0,
                random_state=RANDOM_STATE
            )
        ],
        'models__depth': [4, 6, 8],
        'models__learning_rate': [0.01, 0.03, 0.05],
        'models__l2_leaf_reg': [3, 5, 7, 10],
        'models__iterations': [800, 1200, 1600],
        'preprocessor__num__scaler': [None]
    }]
```

```python
search_cat_boost = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=param_distributions_cat_boost,
    n_iter=20,
    scoring='neg_mean_absolute_error',
    cv=5,
    n_jobs=-1,
    random_state=RANDOM_STATE,
    refit=True,
    verbose=1
)
```

```python
search_cat_boost.fit(X_train, y_train)

print('CatBoostRegressor model and its parameters:\n', search_cat_boost.best_estimator_)
print('CV MAE (CatBoostRegressor):', round(-search_cat_boost.best_score_, 2))
```

```python
param_distributions_hgbr = [
    {
    'models': [HistGradientBoostingRegressor(loss='absolute_error')],
    'models__max_depth': [4, 6, 8],
    'models__learning_rate': [0.01, 0.03, 0.05],
    'models__max_iter': [300, 500, 800],
    'models__l2_regularization': [0.0, 0.1, 1.0],
    'models__min_samples_leaf': [10, 20, 50]
}]
```

## Overall conclusion and recommendations for the client

As part of the project, a machine learning model was developed and trained to predict the final temperature of molten steel during processing. The model is intended for use in a technological process simulation system at the client’s steelworks.

**Object of modelling**: ladle (`key`).

**Target feature**: final state temperature (`final_temperature`).

During the work, seven data sources were analysed, extensive preprocessing and cleaning were performed, correlation analysis and feature engineering were carried out, the datasets were merged, and several models were trained with hyperparameter tuning.

---

<b style="font-size: 18px;">Data preprocessing</b>

Seven data sources were analysed (`data_arc_new.csv`, `data_bulk_new.csv`, `data_bulk_time_new.csv`, `data_gas_new.csv`, `data_temp_new.csv`, `data_wire_new.csv`, `data_wire_time_new.csv`).

- All datasets were cleaned in advance: outliers were removed and missing values were handled.
- Column names were converted to `snake_case`.
- In the `data_arc` dataset, one anomalous outlier in reactive power was removed.
- The feeding time data for bulk and wire materials (`data_bulk_time` and `data_wire_time`) do not contain useful information for predicting the final temperature and were not used in the model.
- The target variable (`final_temperature`) was cleaned of abnormally low values (<1400 °C), which correspond to the onset of steel crystallisation.
- The data in the `data_arc` and `data_bulk` datasets were aggregated by batch (`key`), and all features were summed.
- In the `data_temp` dataset, the target temperature (`final_temperature`) and `initial_temperature` were extracted.

---

<b style="font-size: 18px;">Data merging</b>

- The datasets `data_arc_new.csv`, `data_bulk_new.csv`, `data_gas_new.csv`, `data_temp_new.csv`, and `data_wire_new.csv` were merged by `key`.

---

<b style="font-size: 18px;">Feature engineering</b>


The following new features were introduced:

- `heating_duration_min`
- `apparent_power` (derived from active and reactive power)
- `initial_temperature` and the target `final_temperature`, with intermediate states per batch (`key`) aggregated
- `sum_bulks` — total volume of all bulk materials
- `sum_wires` — total volume of all wire materials
- `total_sum_bulks_wires` — total combined volume of bulk and wire materials

---

<b style="font-size: 18px;">Correlation analysis</b>

During the correlation analysis, multicollinear features and those with no correlation with the target feature were removed (`active_power`, `reactive_power`, `bulk_1`, `bulk_2`, `bulk_3`, `bulk_5`, `bulk_7–9`, `bulk_13`, `wire_3–6`, `wire_8–9`, `total_sum_bulks_wires`).

- `wire_1` (0.30)
- `heating_duration_min` and `bulk_12` (0.24)
- `bulk_4` and `bulk_15` (0.23)
- `sum_bulks` (0.22)
- `apparent_power` (0.17)
- `sum_wires` (0.11)
- `bulk_11` (−0.12)
- `wire_2` (−0.24)
- `wire_6` (−0.26)

For the remaining features (`bulk_10`, `bulk_14`, `wire_7`, `gas_1`), the correlation is very small and almost negligible. However, the model was also trained using these features, although they are potential candidates for removal.

---

<b style="font-size: 18px;">Machine learning model training</b>

The following models were evaluated: `RandomForestRegressor`, `CatBoostRegressor`, and `HistGradientBoostingRegressor`. Hyperparameter tuning was performed within a pipeline using `RandomSearch`.
